```python
material_hogarnacional = pd.read_excel("/content/drive/MyDrive/TP OP/c2022_tp_pob_c1.xlsx")

indices_a_eliminar = [0, 1, 15, 16]
material_hogarnacional = material_hogarnacional.drop(indices_a_eliminar)
material_hogarnacional.columns = ['Material_cubierta_exterior', 'Poblacion_viviendas_particulares', 'pisoprimeracalidad', 'pisomedianacalidad', 'pisomalacalidad', 'pisosindefinir']
material_hogarnacional = material_hogarnacional.reset_index(drop=True)

material_hogarnacional.shape
material_hogarnacional.head(50)

```

```python
#Ahora vamos a tratar de construir con los datos obtenidos de los materiales de las casas en el Censo a nivel nacional una hipotesis acerca de los ingresos de la poblaciòn.
#Para tal fin vamos a sumar las filas y columnas con los materiales de "mejor calidad" tanto en el piso como en la cubierta exterior, contrastandolo con los de menor calidad.
indices_filas = [2, 10]
suma_1 = material_hogarnacional.iloc[indices_filas][['pisoprimeracalidad']].sum()
filas_a_seleccionar = [3, 4, 5, 6, 7, 8, 9, 11, 12]
columnas_a_seleccionar = ['pisomedianacalidad', 'pisomalacalidad']
suma_2 = material_hogarnacional.iloc[filas_a_seleccionar][columnas_a_seleccionar].sum().sum()

#Al no contemplar nuestra base de datos a las personas en situaciòn de calle las agregaremos manualmente con el dato arrojado tambièn por el censo.
total_personas_en_calle = 5705
suma_2 += total_personas_en_calle

# Creamos un nuevo DataFrame con los resultados
status_economico = pd.DataFrame({
    'Mayores_ingresos': [suma_1],
    'Menores_ingresos': [suma_2]
})

# Mostramos el nuevo DataFrame en un gràfico de torta
plt.figure(figsize=(8, 8))
plt.pie(status_economico.iloc[0], labels=status_economico.columns, autopct='%1.1f%%', startangle=140)
plt.title('Ingresos a nivel nacional')
plt.show()


```

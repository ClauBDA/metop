```python
import pandas as pd
import numpy as np
import xlrd
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
import plotly.graph_objs as go
import plotly.express as px
import datetime

from google.colab import files
from google.colab import drive

material_hogarC10 = pd.read_excel("/content/MaterialComuna10.xlsx")
indices_a_eliminar = [0, 1, 15, 16]
material_hogarC10 = material_hogarC10.drop(indices_a_eliminar)
material_hogarC10.columns = ['Material_cubierta_exterior', 'Poblacion_viviendas_particulares', 'pisoprimeracalidadC10', 'pisomedianacalidadC10', 'pisomalacalidadC10', 'pisosindefinirC10']
material_hogarC10 = material_hogarC10.reset_index(drop=True)
material_hogarC10.shape


indices_filasC10 = [2, 10]
suma_101 = material_hogarC10.iloc[indices_filasC10][['pisoprimeracalidadC10']].sum()
filas_a_seleccionarC10 = [3, 4, 5, 6, 7, 8, 9, 11, 12]
columnas_a_seleccionarC10 = ['pisomedianacalidadC10', 'pisomalacalidadC10']
suma_102 = material_hogarC10.iloc[filas_a_seleccionarC10][columnas_a_seleccionarC10].sum().sum()

status_economico_C10 = pd.DataFrame({
    'Mayores_ingresosC10': [suma_101.values[0]],
    'Menores_ingresosC10': [suma_102]
})
print(status_economico_C10)



material_hogarC10.head(45)





```
